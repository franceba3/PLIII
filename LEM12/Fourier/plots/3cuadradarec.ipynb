```python
from numpy import loadtxt
from matplotlib.pyplot import plot, show, xlabel, ylabel
```

$\textbf{1. UN PICO}$

```python
uno=loadtxt(("data/31cuadradareconstruidaunpico.txt"),float,encoding='latin-1', 
            skiprows=2)
t2=uno[:,0]
i2=uno[:,1]
plot(t2,i2)
xlabel("Tiempo (s)")
ylabel("Voltaje (V)")
show()
```
