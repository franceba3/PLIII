```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_excel
from scipy.interpolate import interp1d
```

```python
ib = read_excel("data.xlsx", sheet_name = "Sheet2").to_numpy() #First plot
plt.ylabel("B (mT)")
plt.xlabel("I (A)")
#nlc = lambda x: 1.3559*x**3 - 30.003*x**2 + 243.37*x - 38.966
nll = lambda x: 237.62 * np.log(x) + 197.36
plt.title("Curva de calibración")
plt.plot(np.linspace(0.01, 9.5, 100), nll(np.linspace(0.01, 9.5, 100)), "-m", label = "Curva de calibración")
plt.plot(ib[:,0], abs(ib[:,1]), "k.", label = "Medidas de B")
plt.ylim(0, 725)
plt.xlim(0, 9.2)
plt.legend()
```

```python
#INTERPOLADA
inc = interp1d(ib[:,0], abs(ib[:,1]), kind = "cubic", fill_value = "extrapolate")
plt.ylabel("$B$ (mT)")
plt.xlabel("$I$ (A)")
plt.title("Curva de calibración")
plt.plot(np.linspace(0.01, 9.5, 100), inc(np.linspace(0.01, 9.5, 100)), "-m", label = "Curva de calibración")
plt.plot(ib[:,0], abs(ib[:,1]), "k.", label = "Medidas de B")
plt.ylim(0, 725)
plt.xlim(0, 9.2)
plt.legend()
```

Voltaje $V_H$ contra $B$. Como los valores de corriente para los que se han medido los voltajes no son iguales a los valores para los que se midió $B$, se hará uso de la curva de calibración interpolada.

```python
iv = read_excel("data.xlsx", sheet_name = "Sheet3").to_numpy()
inc(iv[:,0])
iv[:,1] = iv[:,1] * 10
```

```python
#El B que corresponde a cada I es ic(I)
lest = np.polyfit(inc(iv[:,0])*1e-3, iv[:,1], 1)
r = np.corrcoef(inc(iv[:,0])*1e-3, iv[:,1])[0,1]
print("Regresión: {}x + {}. R^2 = {}".format(lest[0], lest[1], r**2))
plt.plot(inc(iv[:,0])*1e-3, lest[0] * inc(iv[:,0])*1e-3 + lest[1], "--r", label = "Ajuste Lineal")
plt.plot(inc(iv[:,0])*1e-3, iv[:,1], "k.", label = "Valores experimentales")
plt.text(0.45, -7, "$V_H$ = {}$B$ -{}\n R$^2$ = {}".format(round(lest[0], 3), abs(round(lest[1], 3)), round(r**2, 3)))
plt.title("Ajuste lineal $V_H$ ~ $B$")
plt.xlabel("$B$ (T)")
plt.ylabel("$V_H$ ($\mu$V)")
plt.legend()
```
